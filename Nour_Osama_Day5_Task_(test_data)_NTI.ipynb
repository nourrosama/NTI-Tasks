```python
import pandas as pd
test_data=pd.read_csv('test_data.csv')
test_data.head()
```

# Preprocessing

```python
#Import the necessary modules:
import pandas as pd
import string

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = str(text).lower()                        # lowercase
    text = re.sub(r'http\S+|www\S+', '', text)      # remove urls
    text = re.sub(r'@\w+', '', text)                # remove mentions
    text = re.sub(r'#\w+', '', text)                # remove hashtags
    text = re.sub(r'[^a-z\s]', '', text)            # remove punctuation/numbers
    words = text.split()
    words = [w for w in words if w not in stop_words]  # remove stopwords
    return " ".join(words)

test_data['cleaned_test_data'] = test_data['Phrase'].apply(preprocess)
test_data[['Phrase','cleaned_test_data']].head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize, word_tokenize
#to tokenize all records
for text in test_data['cleaned_test_data']:
    print(sent_tokenize(text))
    print(word_tokenize(text))
    print()
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer
nltk.download('punkt')
porter = PorterStemmer()
snowball = SnowballStemmer("english")

text = test_data['cleaned_test_data'][0]
words = word_tokenize(text)
print("Word\t\tPorter\t\tSnowball")
print("-" * 40)
for word in words:
    print(f"{word}\t\t{porter.stem(word)}\t\t{snowball.stem(word)}")
```

```python
X = test_data['cleaned_test_data']
y = test_data['target']
```

```python
def predict(x,model):
    pred=model.predict(x)
    return pred
```

```python
import pickle
with open(r"clf_model.pkl", 'rb') as f:
    vc_clf = pickle.load(f)
```

```python
predictions_clf = []

for index in range(len(X)):
    row_text = X.iloc[[index]]
    predictions_clf.append(predict(row_text, vc_clf)[0])
```

```python
from sklearn import metrics

print(metrics.confusion_matrix(y,predictions_clf))

print(metrics.classification_report(y,predictions_clf))

print(metrics.accuracy_score(y,predictions_clf))
```
