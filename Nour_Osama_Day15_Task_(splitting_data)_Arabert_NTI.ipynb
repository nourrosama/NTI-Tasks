```python
import pandas as pd
```

```python
data = pd.read_csv("/content/Train.xlsx - Sheet1 (1).csv", encoding="utf-8")
print(data.head())
```

```python
data.columns
```

```python
rows_per_class = 1000
data2 = []

# Take only the first 5 unique categories
for category in data['category'].unique()[:5]:
    category_df = data[data['category'] == category]
    sampled_category_df = category_df.sample(
        n=rows_per_class, replace=True, random_state=42
    )
    data2.append(sampled_category_df)

data2 = pd.concat(data2)

print(data2["category"].value_counts())
print(data2.shape)
```

```python
# Save to CSV
data2.to_csv("data2.csv", index=False, encoding="utf-8-sig")
```

```python
data2.head()
```

```python
X = data2[['q_body']]
y = data2['category']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

print('train shape:', X_train.shape)
print('test shape: ', X_test.shape)
```

```python
train_data = pd.DataFrame(X_train)
train_data['target'] = y_train

# Now train_data contains both features and the target
print(train_data.head())
```

```python
train_data.head()
```

```python
train_data.to_csv('train_data.csv')
```

```python
test_data = pd.DataFrame(X_test)
test_data['target'] = y_test

# Now train_data contains both features and the target
print(test_data.head())
```

```python
test_data.to_csv('test_data.csv')
```
